```python
sst_day = ds_hourly['SST'].groupby('time_counter.day').mean().values*sea_mask_modified_nan
T2 = ds_hourly['T2'].groupby('time_counter.day').mean().values*sea_mask_modified_nan
q2 = ds_hourly['Q2'].groupby('time_counter.day').mean().values*sea_mask_modified_nan
pblh = ds_hourly['PBLH'].groupby('time_counter.day').mean().values*sea_mask_modified_nan
```

```python
s = 100   # 'inf'

dsst_day = np.zeros(sst_day.shape)
dT2 = dsst_day.copy()
dq2 = dsst_day.copy()
dpblh = dsst_day.copy()


sst_ave, T2_ave, q2_ave, pblh_ave = \
np.zeros(sst_day.shape), np.zeros(sst_day.shape), np.zeros(sst_day.shape), np.zeros(sst_day.shape)

for t in range(0,sst_day.shape[0]):
    if t % 50 == 0:
        print(t)
    
    sst_ave[t] = gm.nan_gaussian_filter(sst_day[t],s)                             
    dsst_day[t] = sst_day[t] - sst_ave[t]
    
    T2_ave[t] = gm.nan_gaussian_filter(T2[t],s)                             
    dT2[t] = T2[t] - T2_ave[t]
    
    q2_ave[t] = gm.nan_gaussian_filter(q2[t],s)                             
    dq2[t] = q2[t] - q2_ave[t]
    
    pblh_ave[t] = gm.nan_gaussian_filter(pblh[t],s)                             
    dpblh[t] = pblh[t] - pblh_ave[t]
```
